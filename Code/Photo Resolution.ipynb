```python
from PIL import Image, ImageDraw
def visualize(image_path):
    img = Image.open(image_path).convert("RGB")
    print(f"Orignal Photo Resolution: {img.size}")
    configs = [(512, 32),(512, 64),(512, 128)]
    for target_size, grid_size in configs:
        resized = img.resize((target_size, target_size), Image.LANCZOS)
        center = target_size // 2
        crop_size = 200
        cropped = resized.crop((
            center - crop_size,
            center - crop_size,
            center + crop_size,
            center + crop_size
        ))
        cropped = cropped.resize((800, 800), Image.NEAREST)
        draw = ImageDraw.Draw(cropped)
        scale = 800 / (crop_size * 2)
        grid_scaled = int(grid_size * scale)
        for x in range(0, 800, grid_scaled):
            draw.line([(x, 0), (x, 800)], fill=(255, 0, 0), width=2)
        for y in range(0, 800, grid_scaled):
            draw.line([(0, y), (800, y)], fill=(255, 0, 0), width=2)
        out_path = f"grid_{target_size}px_{grid_size}cell_zoom.jpg"
        cropped.save(out_path)
        print(f"✓ {target_size}x{target_size}, grid {grid_size}px → {out_path}")
```

```python
if __name__ == "__main__":
    visualize("../HDR_20260223_105400.jpg")
```
